```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF, LatentDirichletAllocation
from gensim.utils import simple_preprocess

nltk.download('punkt')
nltk.download('stopwords')

# Load dataset
df = pd.read_excel('Assignment.xlsx')

# Preprocessing function
def clean_article(article):
    tokens = word_tokenize(article)
    tokens = [token.lower() for token in tokens]
    tokens = [re.sub(r'[^a-zA-Z0-9]', '', token) for token in tokens if token.isalnum()]
    stop_words = set(stopwords.words('english'))
    cleaned_tokens = [token for token in tokens if token not in stop_words]
    return ' '.join(cleaned_tokens)

df['Cleaned_Article'] = df['Article'].apply(clean_article)
```

```python
tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')
tfidf = tfidf_vectorizer.fit_transform(df['Cleaned_Article'])
```

```python
num_topics = 5  # Adjust as needed
nmf_model = NMF(n_components=num_topics, random_state=42)
nmf_topics = nmf_model.fit_transform(tfidf)

# Function to get top words for each topic
def get_top_words(model, feature_names, n_top_words):
    top_words = {}
    for topic_idx, topic in enumerate(model.components_):
        top_words[topic_idx] = [feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]
    return top_words

tfidf_feature_names = tfidf_vectorizer.get_feature_names_out()
top_words_nmf = get_top_words(nmf_model, tfidf_feature_names, 10)
```

```python
topic_labels = {topic: " ".join(words[:3]) for topic, words in top_words_nmf.items()}

# Print topic labels
for topic_id, label in topic_labels.items():
    print(f"Topic {topic_id}: {label}")
```

```python
# Assign themes to articles based on the dominant topic
article_themes = []
for i in range(len(df)):
    topic = nmf_topics[i].argmax()
    theme = topic_labels[topic]
    article_themes.append(theme)

# Print themes for each article
for i, theme in enumerate(article_themes):
    print(f"Article {i+1} - Theme: {theme}")
```

```python
import streamlit as st
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from textblob import TextBlob
from bertopic import BERTopic
import nltk

nltk.download('punkt')
nltk.download('stopwords')

# Load your dataset
df = pd.read_excel('Assignment.xlsx')

# Clean Up Articles Function
def clean_article(article):
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(article)
    filtered_article = [w for w in word_tokens if w.isalnum() and not w in stop_words]
    return " ".join(filtered_article)

df['Cleaned_Article'] = df['Article'].apply(clean_article)

# Check the Mood Function
def get_mood(article):
    analysis = TextBlob(article)
    if analysis.sentiment.polarity > 0:
        return "Positive"
    elif analysis.sentiment.polarity < 0:
        return "Negative"
    else:
        return "Neutral"

df['Mood'] = df['Cleaned_Article'].apply(get_mood)

# Find Connections Function using BERTopic
model = BERTopic()
topics, probabilities = model.fit_transform(df['Cleaned_Article'].tolist())

# Adding topics to DataFrame
df['Topic'] = topics
topic_labels = model.get_topic_info()
df['Detailed_Theme'] = df['Topic'].apply(lambda x: topic_labels.iloc[x + 1]['Name'])

# Streamlit App
st.title('News Analysis Project: Uncover What Matters')

st.sidebar.title("Options")
options = st.sidebar.selectbox("Choose an analysis type", ["Cleaned Articles", "Mood Rating", "Detailed Themes"])

if options == "Cleaned Articles":
    st.header("Cleaned Articles")
    for i, row in df.iterrows():
        st.subheader(f"Article {i+1}")
        st.write(row['Cleaned_Article'])

elif options == "Mood Rating":
    st.header("Mood Rating")
    for i, row in df.iterrows():
        st.subheader(f"Article {i+1} - Mood: {row['Mood']}")
        st.write(row['Cleaned_Article'])

elif options == "Detailed Themes":
    st.header("Detailed Themes")
    for i, row in df.iterrows():
        st.subheader(f"Article {i+1} - Theme: {row['Detailed_Theme']}")
        st.write(row['Cleaned_Article'])
```
